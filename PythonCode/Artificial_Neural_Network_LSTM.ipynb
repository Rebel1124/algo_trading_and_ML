```python
# 'Backcandle' Setup for 'data_set_scaled' X-Feature Processing & 'y' Target Setting 

# Create empty list to hold backcandle created arrays for target window processing
x = []

# Set backcandle variable to backtest number of backcandle/past data X features to be iterated each LSTM recursive loop
backcandles = 30

# Display/preview 'data_set_scaled' shape[0] (i.e. column numbers, or potential X-Features)
display(f'data_set_scaled post-MinMaxScaler shape: {data_set_scaled.shape[0]}')


for j in range ((data_set_scaled.shape[1]-3)):
    x.append([])
    #display(f'Empty array instances: {x}')
    for i in range (backcandles, data_set_scaled.shape[0]): # iterate in for loop through every index row in the data_set_scaled array
        x[j].append(data_set_scaled[i-backcandles:i, j])
        
# Reshape array by moving axis from position 0 to position 2 (i.e x-->z)
x = np.moveaxis(x, [0], [2])

# Remove/resize 'y' to remove the first 'backcandles' since backtesting won't include them
# Re-assign X to np.array format
# Choose target column as 'Target_Next_Close' for 'y' (or [backcandles:, -1], where -1 == 'Target_Next_Close'

x, y_temp = np.array(x), np.array(x_inversed[backcandles:, -2])

# Reshape y_temp to vertical column vectors for ML processing format
y = np.reshape(y_temp, (len(y_temp), 1))

# Print preview the 'x' & 'y' final processed array shapes

# Preview 'x' shape
display(f'x shape: {x.shape}')
display(f'y shape: {y.shape}')
print()
display(f'x shape[0] (index rows) = {x.shape[0]}.')
display(f'x shape[1] (backcandle dim) = {x.shape[1]}.')
display(f'x shape[2] (number of X-features used) = {x.shape[2]}.')
display(f'Shape of data_set_scaled should be {data_set_scaled.shape[0]} - {backcandles} = {x.shape[0]}.')
```

## <i> iii.) Constructing the LSTM Model Using Sequential Model & LSTM Module</i>

```python
# Insantiate the LSTM Neural Network Model
import tensorflow as tf 
import keras 
from keras import optimizers
from keras.callbacks import History
from keras.models import Model
from keras.layers import Dense, Dropout, LSTM, Input, Activation, concatenate
import numpy as np


# Create model instance variables
lstm_input = Input(shape=(backcandles, (data_set_scaled.shape[1]-3)), name='lstm_input')
inputs = LSTM(30, name='first_layer')(lstm_input) # default was set to 30 
inputs = Dense(1, name='dense_layer')(inputs)
outut = Activation('linear', name='output')(inputs)
nn = Model(inputs=lstm_input, outputs=outut)
adam = optimizers.Adam()

# Compile LSTM Model
nn.compile(optimizer=adam, loss='mse', metrics=['accuracy'])

# Fit LSTM Model
fit_model = nn.fit(x=X_train, y=y_train, batch_size=128, epochs=200, shuffle=True)
```

```python
# Plot the Output of the fit_model history
fig, ax = plt.subplots(1) # Creates figure fig and add an axes, ax.
fig2, ax2 = plt.subplots(1) # Another figure

ax.set_title(f' ANN LSTM ((Linear Output Act.) - {algoData_asset} Accuracy Plot')
ax.plot(fit_model.history['accuracy'], label='train')
fig.savefig(f'../Datasets/algoData_LSTM_results/{algoData_asset}_LSTM_accuracy_.png')


ax2.set_title(f' ANN LSTM (Linear Output Act.) - {algoData_asset} Loss Plot')
ax2.plot(fit_model.history['loss'], label='train')
fig2.savefig(f'../Datasets/algoData_LSTM_results/{algoData_asset}_LSTM_loss_.png')
```

```python
# Evaluate the model using testing data
model_loss, model_accuracy = nn.evaluate(X_train, y_train, verbose=2)

# Display evaluation results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Create function to compile cumulative returns for Artificial Neural Network Predicted y_test values

def annReturns(results, X_test):

    # Make emtpy dataframe to store actual returns, predictions & cumulative returns
    
    # Inititialize first return 0 so cumulative ret begins at base 1
    #annData['Actual Returns'][0] = 0
    
    annData = pd.DataFrame()
    annData.index = data_set_concat.index
    #results.index = X_test.index
    annData['Prediction_Signal'] = data_set_concat['y_pred']
    annData['Actual_Returns'] = data_set_concat['ActualReturns'].values
    annData['annStrategyReturns'] = annData['Actual_Returns'] * annData['Prediction_Signal'] 
    annData['cumActual_Returns'] = (1 + annData['Actual_Returns']).cumprod()
    annData['cumANNStrategyReturns'] = (1 + annData['annStrategyReturns']).cumprod()
    
    returns = annData
    
    return returns
```

```python
# Run annReturns Function and display sample final cumulative returns dataframe

returns_df = annReturns(results, X_test)
display(returns_df.head(5))
display(returns_df.tail(5))
```
